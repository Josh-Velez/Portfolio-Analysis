```python
import pandas as pd 
import numpy as np
import yfinance as yf
import datetime as dt
```

```python
#historical data begins 10 years from today
#more efficient code to write "period = 10y" instead of "start date"?

assets = ['GE', 'RTX', 'LMT', 'BA', 'GD', 'SWBI', 'NOC']
benchmarks = ['spy', 'iwm', 'dia']
startdate= '2014-09-24'
tickers = assets + benchmarks

benchmarks = [asset.upper() for asset in benchmarks]
assets = [asset.upper() for asset in assets]
data = yf.download(tickers, start=startdate)['Adj Close']


#divide by 10 to calculate trading days for previous 10 years
len(data)/10
```

Annualized Volatility

```python
data[assets].pct_change(252).std()
asset_volatility = data[assets].pct_change(252).std()
asset_volatility
```

```python
data[assets].pct_change(252).mean()
asset_return = data[assets].pct_change(252).mean()
asset_return
```

```python
data[benchmarks].pct_change(252).mean()
market_returns = data[benchmarks].pct_change(252).mean()
market_returns

```

```python
data[benchmarks].pct_change(252).std()
market_volatility = data[benchmarks].pct_change(252).std()
market_volatility
```

```python
#calculate total returns using trailing 10 years
#Formula = (1 + Stock Rts).cumprod()
#https://site.financialmodelingprep.com/developer/docs/blog/portfolio-risk-and-returns-with-python 

total_returns = (1 + asset_return).cumprod()
total_returns

#https://www.w3schools.com/python/pandas/ref_df_cummax.asp

#cumulaitive max function

drawdown = data[assets]
average_weekly_drawdown = drawdown.pct_change(252).cummax().mean()
average_weekly_drawdown

max_weekly_drawdown = drawdown.pct_change(252).max()
max_weekly_drawdown
```

Beta against benchmarks. 
Beta formula is Covariance(Stock Returns, Market Returns)/Variance(Market Returns)

https://www.codingfinance.com/post/2018-04-05-portfolio-returns-py/


Beta = (return of a stock + return of the index)/ variance of the index


```python
#Create a table showing portfolio (assets) risk analysis in equal weighted portfolio analysis as of current date

table1 = pd.DataFrame(index=assets)

table1['Portfolio Weight'] = 1/len(assets)
table1['Annualized Volatility'] =  asset_volatility
table1['Beta against SPY'] = (asset_return + market_returns['SPY'])/market_volatility['SPY']
table1['Beta against IWM'] = (asset_return + market_returns['IWM'])/market_volatility['IWM']
table1['Beta against DIA'] = (asset_return + market_returns['DIA'])/market_volatility["DIA"]
table1['Avg Weekly Drawdown'] = average_weekly_drawdown
table1['Max Weekly Drawdown'] = max_weekly_drawdown
table1['Total Return'] = total_returns
table1['Annualized Total Return'] = asset_return

table1
```

```python
#sharpe ratio
#https://www.quantifiedstrategies.com/how-to-calculate-the-sharpe-ratio-in-python/#Calculating_the_Sharpe_Ratio_for_S_P_500_and_Russell_2000 

df = pd.DataFrame(index=assets)

df['asset_return'] = asset_return
df['asset vol'] = asset_volatility
df['portfolio return'] = df.mean(axis=1)
df['portfolio volatility'] = df.std(axis=1)
df['sharpe ratio'] = df['portfolio return']/df['portfolio volatility']

#https://python.plainenglish.io/analyzing-etf-flows-and-correlations-with-the-s-p-500-python-streamlit-guide-fbe3d672c277

#corr.() function

#tracking errors
#https://www.investopedia.com/terms/t/trackingerror.asp#:~:text=Given%20a%20sequence%20of%20returns,and%20B%20is%20benchmark%20return.

df['portfolio return']

total_port_return = df['portfolio return'].sum()

total_port_return

df
```

```python
#Create a table showing Portfolio Risk against ETFS

ETF = ['SPY', 'IWM', 'DIA']

etf_data = yf.download(ETF, start=startdate)

returns = etf_data['Adj Close'].pct_change(252)

correlation_against_etf = returns.corr()
covariance = returns.cov()
sharpe = returns.mean()/returns.std()
vol = (returns.std()- returns['SPY'].std()).abs()


table3 = pd.DataFrame(index=benchmarks)
table3['TIcker'] = ETF

table3['Correlation'] = correlation_against_etf['SPY']
table3['Covariance'] = covariance['SPY']
table3['Sharpe Ratio'] = sharpe[ETF]
table3['Tracking Errors'] = (total_port_return - returns['SPY']).std()
table3['Annualized Volatility'] = vol[ETF]

table3


```
